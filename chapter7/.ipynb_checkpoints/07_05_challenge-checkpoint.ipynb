```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
pd.options.display.max_rows = 10
```

```python
allyears = pd.read_csv('allyears.csv.gz').dropna()
allyears.info()
```

```python
match_names = list(np.intersect1d(allyears[(allyears.sex == 'F')].name.unique(),
                                  allyears[(allyears.sex == 'M')].name.unique()))
len(match_names)
```

```python
female_names = allyears[(allyears.sex == 'F') & (allyears.name.isin(match_names))].groupby(['name', 'sex']).sum()
male_names = allyears[(allyears.sex == 'M') & (allyears.name.isin(match_names))].groupby(['name', 'sex']).sum()
```

```python
merged_names = female_names.merge(male_names, on = 'name', suffixes = ['_f', '_m'], how = 'inner')
```

```python
merged_names['m_to_f_ratio'] = merged_names.number_m / merged_names.number_f
merged_names['total_count'] = merged_names.number_m + merged_names.number_f
merged_names.head(5)
```

```python
top_ten_unisex = list(merged_names[(merged_names.m_to_f_ratio > 0.5) & (merged_names.m_to_f_ratio < 2.0)].sort_values('total_count', ascending = False).head(10).reset_index().name)
top_ten_unisex
```

```python
def plotname(sex, name):
    data = allyears.query('sex == @sex and name == @name')
    
    pp.plot(data.year, data.number, label=name)
    pp.axis(xmin=1880, xmax=2018)
```

```python
pp.figure(figsize=(12,2.5))

for name in top_ten_unisex:
    plotname('F', name)

pp.legend()
```

```python
pp.figure(figsize=(12,2.5))

for name in top_ten_unisex:
    plotname('M', name)

pp.legend()
```

```python
allyears_indexed = allyears.set_index(['sex', 'name', 'year']).sort_index()
pp.figure(figsize=(9,9))

# loop over top-ten unisex names
for i, name in enumerate(top_ten_unisex):
    pp.subplot(5,2,i+1)

    pp.plot(allyears_indexed.loc['M',name], label='M')
    pp.plot(allyears_indexed.loc['F',name], label='F')
    
    pp.legend()
    pp.title(name)

# improve subplot spacing
pp.tight_layout()
```
